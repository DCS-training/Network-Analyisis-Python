# Demo 3

```python
pubdates = np.array(df.publication_date)
langs = df["language"]
langs_list = list(langs)
print(type(pubdates))
print(type(langs))
print(type(langs_list))
```

```python
print("Earliest publication date: ", pubdates.min())
print("Latest publications date:  ", pubdates.max())
```

```python
langs.value_counts()
```

```python
langs.unique()
```

```python
df_grouped = df.groupby('language').count()
df_grouped.drop(columns=["author", "topic", "publication_place", "publication_date"], inplace=True)
```

```python
df_grouped
```

```python
df_grouped.sort_values("title")
```

```python
clean_langs = []
for lang in langs:
    cleaned = lang.replace(".", "")
    cleaned = lang.strip()
    clean_langs += [cleaned]
clean_langs = pd.Series(clean_langs)
print("Before: ", len(langs.unique()))
print("After:  ", len(clean_langs.unique()))
```

```python
df['language'] = clean_langs
# df.language = clean_langs
```

```python
df_grouped = df.groupby('language').count()
df_grouped.drop(columns=["author", "topic", "publication_place", "publication_date"], inplace=True)
```

```python
df_grouped.sort_values("title", inplace=True)
df_grouped
```
